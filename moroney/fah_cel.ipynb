```
import tensorflow as tf
import numpy as np
```

```
print(tf.__version__)
```

```
celsius = np.array([-40,-10,0,8,15,22,38])
fahrenheit = np.array([-40,14,32,46,59,72,100])
```

```
for index, value in enumerate(celsius):
  print('{} degree celsius = {} degree fahrenheit'.format(value, fahrenheit[index])) 
```

```
type(celsius)
```

```
model = tf.keras.Sequential(
    [
     tf.keras.layers.Dense(1, input_shape=[1])
    ]
)
```

```
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.mean_squared_error,
)
```

```
history = model.fit(celsius, fahrenheit, epochs=500, verbose=1)
```

```
import matplotlib.pyplot as plt
% matplotlib inline
```

```
plt.plot(history.history['loss'])
plt.show()
```

```
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.1),
    loss = tf.keras.losses.mean_squared_error,
)
```

```
history = model.fit(celsius, fahrenheit, epochs=500, verbose=False)
```

```
plt.plot(history.history['loss'])
plt.show()
```

```
pred = model.predict([100.0])
pred
```

```
model.get_weights()
```

```
model = tf.keras.Sequential(
    [
     tf.keras.layers.Dense(4, input_shape=[1]),
     tf.keras.layers.Dense(4),
     tf.keras.layers.Dense(1),
    ]
)
```

```
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.mean_squared_error,
)
```

```
history = model.fit(celsius, fahrenheit, epochs=500, verbose = 1)
```

```
plt.plot(history.history['loss'])
```

```
model.predict([100.0])
```

```
layer_0 = tf.keras.layers.Dense(units=1, input_shape=[1])
model_1 = tf.keras.Sequential([layer_0])
model_1.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.1), loss='mean_squared_error')
history = model_1.fit(celsius, fahrenheit, epochs=500, verbose=1)
model_1.predict([100.0])
```

```
layer_0.get_weights()
```

```
l_0 = tf.keras.layers.Dense(4, input_shape=[1])
l_1 = tf.keras.layers.Dense(4)
l_2 = tf.keras.layers.Dense(1)
model = tf.keras.Sequential([l_0, l_1, l_2])
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.1),
    loss = 'mean_squared_error'
)
history = model.fit(celsius, fahrenheit, epochs=500, verbose = 1)
model.predict([100.0])
```
